```python
import os
import numpy as np
import matplotlib.pyplot as plt
import pytvfemd
```

```python
def plot(x1, x2, s):
    plt.figure(figsize=(10, 10))
    for i in range(s):
        plt.subplot(int(s / 3) + s % 3, 3, i + 1)
        plt.plot(x1[:, i], 'k-')
        plt.plot(x2[:, i], 'r--', linewidth=2)
    plt.show()
```

```python
def test_sin():
    m_sin = np.loadtxt(os.path.join("test_files/matlab_sin.txt"))
    x = 0.3 * np.sin(2.0 * np.pi * np.arange(0, 1000, dtype=int) / 365.0)
    p_sin = pytvfemd.tvfemd(x)
    
    plot(m_sin, p_sin, m_sin.shape[1])
```

```python
test_sin()
```

```python
def test_ts1():
    m_ts1 = np.loadtxt("test_files/matlab_ts1.txt")
    x = np.loadtxt("test_files/ts1.txt")
    p_ts1 = pytvfemd.tvfemd(x)
    
    plot(m_ts1, p_ts1, m_ts1.shape[1])
```

```python
test_ts1()
```
